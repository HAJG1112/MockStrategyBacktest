```python
import pandas as pd
from dateutil import relativedelta
from engine import OrtexBacktestEngine
df = pd.read_csv('../Pdata/data.csv', index_col = 0, parse_dates = True)
df.reset_index(inplace = True)
# remap the buy and sell to a binary output
df.buy = df.buy.astype(int)
df.signal_return = df.signal_return.astype(float)
df_long = df.loc[df.buy == 1]
df_long['exit_date'] = df_long['date'] + pd.to_timedelta(df_long.rec_holding_days, unit = 'd')
df_long.set_index(['date', 'ticker'], inplace = True)

# to create an algorithm structure free of a trading startegy. We will create a portfolio class
# in here, when we select a trade, we will store the ticker, the signal return and the exit date. 
# checking the current date against the exit date, we map the returns to the amount invested in the security
# calculate the new amount, and then add this position back to our cas#h position
# we will enter new trades in the following day

bt = OrtexBacktestEngine(df_long, 100000)
bt.run(80)
```
